##Initial Attempt: Load Model from HF

The model used in the paper for the poem reversal task was LLaDA-8B-Instruct.

The examples they provided in the paper were as follows:

Example 1:  
Prompt: 窈窕淑女的下一句是什么？直接输出句子即可。

Example 2:  
Prompt: 不拘一格降人才的上一句是什么？直接输出句子即可。  
Answer: 我劝天公重抖擞。

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

model_name = "GSAI-ML/LLaDA-8B-Instruct"

tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    trust_remote_code=True,
    device_map="auto",
    torch_dtype=torch.bfloat16
)
model.eval()
```

```python
data = [
  {
    "prompt": "窈窕淑女的下一句是什么？直接输出句子即可。",
    "answer": "君子好逑。"
  },
  {
    "prompt": "不拘一格降人才的上一句是什么？直接输出句子即可。",
    "answer": "我劝天公重抖擞。"
  }
]
```

```python
def normalize(text):
  text = text.strip()
  text = text.replace(" ", "")
  text = text.replace("？", "").replace("。", "")
  text = text.replace(":", "").replace("：", "")
  return text
```

```python
def build_prompt(prompt):
  messages = [
    {"role": "user", "content": prompt}
  ]
  return tokenizer.apply_chat_template(
    messages,
    tokenize=False,
    add_generation_prompt=True
  )
```

```python
def generate(prompt):
  prompt = build_prompt(prompt)
  inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

  with torch.no_grad():
    output = model.generate(
      **inputs,
      max_new_tokens=16,
      use_cache=False,
      do_sample=True,
      pad_token_id=tokenizer.eos_token_id
    )

  generated_tokens = output[0][inputs["input_ids"].shape[-1]:]
  decoded = tokenizer.decode(generated_tokens, skip_special_tokens=True)
  return decoded.strip()
```

```python
def evaluate(data):
  correct = 0

  for item in data:
    pred = generate(item["prompt"])
    desired = item["answer"]

    print("PROMPT:", item["prompt"])
    print("MODEL OUTPUT:", pred)
    print("PAPER OUTPUT", desired)
    print("-" * 40)

    if normalize(pred) == normalize(desired):
      correct += 1

  return correct / len(data)
```

```python
acc = evaluate(data)
print("Accuracy:", acc)
```
